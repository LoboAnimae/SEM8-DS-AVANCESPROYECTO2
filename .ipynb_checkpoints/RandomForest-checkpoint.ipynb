```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, plot_roc_curve
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('./vaccination-data.csv')
data.head()
```

```python
data.dtypes
```

```python
data.dropna(subset=['new_deaths_per_million'], inplace=True)
```

```python
label_encoder = LabelEncoder()
data['iso_code'] = label_encoder.fit_transform(data['iso_code'])
```

```python
X = data[['date', 'iso_code', 'new_cases_per_million', 'new_vaccinations', 'total_vaccinations', 'population_density', 'median_age']]
y = data['new_deaths_per_million']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```
